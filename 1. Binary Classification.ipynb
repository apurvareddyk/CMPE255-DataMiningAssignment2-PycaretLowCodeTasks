```python
!kaggle datasets download -d redwankarimsony/heart-disease-data
!unzip heart-disease-data.zip
import pandas as pd

data = pd.read_csv('heart_disease_uci.csv')
```

```python
data = data.drop(columns=['dataset'])
```

```python
print(data.dtypes)
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
data['sex'] = le.fit_transform(data['sex'])
data['cp'] = le.fit_transform(data['cp'])
data['restecg'] = le.fit_transform(data['restecg'])
data['slope'] = le.fit_transform(data['slope'])
data['ca'] = le.fit_transform(data['ca'])
data['thal'] = le.fit_transform(data['thal'])
```

```python
#Using functional API

from pycaret.classification import *
s = setup(data, target = 'num', session_id = 123,use_gpu=True)
```

```python
#using OOP

from pycaret.classification import ClassificationExperiment
exp = ClassificationExperiment()
```

```python
type(exp)
```

```python
exp.setup(data, target = 'num', session_id = 123, use_gpu=True)
```

```python
# comparing models
best = compare_models()
```

```python
# using OOP to compare models
exp.compare_models()
```

```python
plot_model(best, plot = 'confusion_matrix')
```
